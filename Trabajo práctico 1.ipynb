```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

**Ejercicio 1**. Una onda transversal se propaga en una cuerda con una velocidad de 10 m/s y a una longitud de onda de 2 m. Calcular la frecuencia y el per√≠odo de la onda.

onda transversal.png

```python
# C√≥digo de resoluci√≥n ejercicio 1
v = 10  # Velocidad de propagaci√≥n (m/s)
lambda_wave = 2  # Longitud de onda (m)

f = v / lambda_wave  # Frecuencia (Hz)
T = 1 / f  # Per√≠odo (s)

# Mostrar ambas ecuaciones en una misma l√≠nea
display(Math(f"f = \\frac{{v}}{{\\lambda}} = \\frac{{{v}}}{{{lambda_wave}}} = {f:.2f}\\text{{ Hz}}"))
print ('')
display(Math(f"T = \\frac{{1}}{{f}} = \\frac{{1}}{{{f:.2f}}} = {T:.4f}\\text{{ s}}"))

```

**Ejercicio 2.** Considerando que la luz visible tiene una longitud de onda de aproximadamente 500 nm, calcular su frecuencia en el vac√≠o.
espectro_em.png

```python
c = 3.0e8  # Velocidad de la luz (m/s)
lambda_light = 500e-9  # Longitud de onda (m)
f_light = c / lambda_light  # Frecuencia de la onda electromagn√©tica (Hz)

display(Math(f"f_{{light}} = \\frac{{c}}{{\\lambda}} = \\frac{{{c}}}{{{lambda_light}}} = {f_light:.2e}\\text{{ Hz}}"))
```

**Ejercicio 3.** Representar gr√°ficamente la forma de la onda en funci√≥n de la posici√≥n y el tiempo.

```python
x = np.linspace(0, 4 * lambda_wave, 500)  # Posici√≥n en la cuerda (m)
t = 0  # Tiempo inicial
y = np.sin(2 * np.pi * (x / lambda_wave - t / T))  # Ecuaci√≥n de la onda

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Onda transversal en la cuerda')
plt.xlabel('Posici√≥n (m)')
plt.ylabel('Desplazamiento (m)')
plt.title('Representaci√≥n de una onda transversal')
plt.legend()
plt.grid()
plt.show()
```
